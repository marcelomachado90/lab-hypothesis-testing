**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
# H0: mu HP grass = mu HP dragon
# H1: mu HP grass != mu HP dragon

alpha = 0.05

df_dragon = df[(df["Type 1"]=="Dragon") | (df["Type 2"]=="Dragon")]
df_grass = df[(df["Type 1"]=="Grass") | (df["Type 2"]=="Grass")]

st.ttest_ind(df_dragon["HP"], df_grass["HP"])

```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#code here
df_legendary = df[df["Legendary"]==True]
df_non_legendary = df[df["Legendary"]==False]

stats = ["HP", "Attack", "Sp. Atk", "Sp. Def", "Speed"]
alpha = 0.05

for stat in stats:
  print("----------------------------------------")
  print(f"Performing Independence Test for {stat}")
  p_value = st.ttest_ind(df_legendary[stat], df_non_legendary[stat])[1]
  if p_value < alpha:
    print("We can reject the null hypothesis", p_value)
    print(f"Sample mean for legendary {df_legendary[stat].mean()}")
    print(f"Sample mean for non legendary {df_non_legendary[stat].mean()}")
  else:
    print("We did not gather enough evidence to reject the null", p_value)

#Here we didnt really print t_stat because we rather look at p_value which gives a probabililty of second sample belonging to the same
#distribution as the first one, while t_stat gives you "how many std deviations" second sample needed to be away from the first one
```

**Challenge 2**

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 3.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.


```python
def get_distance_school(row):
  school = (-118, 37)
  return np.sqrt((row["longitude"] - school[0])**2 + (row["latitude"] - school[1])**2)

def get_distance_hospital(row):
  hospital = (-122, 34)
  return np.sqrt((row["longitude"] - hospital[0])**2 + (row["latitude"] - hospital[1])**2)

df["distance_school"] = df.apply(get_distance_school, axis=1)
df["distance_hospital"] = df.apply(get_distance_hospital, axis=1)

df.head()
```

```python
df["close?"] = df.apply(lambda x: min(x["distance_school"], x["distance_hospital"]) < 3, axis=1)
df.head()
```

```python
df_close = df[df["close?"]==True]
df_far = df[df["close?"]==False]
```

```python
# H0: mu Price close = mu Price far
# H1: mu Price close != mu Price far

alpha = 0.05

st.ttest_ind(df_close["median_house_value"], df_far["median_house_value"], equal_var = False)
```

```python
# now we believe that houses close are actually more expensive

# H0: mu Price close <= mu Price far
# H1: mu Price close > mu Price far

st.ttest_ind(df_close["median_house_value"], df_far["median_house_value"], alternative = "greater",equal_var = False)
```
